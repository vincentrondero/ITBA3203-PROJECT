```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
sbRate = pd.read_csv("datasets/sbRates.csv", skiprows=1)
sbRate.drop(sbRate[sbRate['Uncertainty.Bounds*'] == 'Lower'].index, inplace=True)   #data cleaning: removing the lower bounds of the data
sbRate.drop(sbRate[sbRate['Uncertainty.Bounds*'] == 'Upper'].index, inplace=True)   #data cleaning: removing the upper bounds of the data

sbRates = sbRate.dropna(axis=1, how='all').drop(sbRate.columns[1:12], axis=1)       #data cleaning: removing columns with NaN values
sbRates.columns = sbRates.columns.str.rstrip('.5')                                  #renaming the column headers

sbRates.rename(columns={'201': '2015', 'Region.Name': 'Region Name'}, inplace=True) #renaming columns

sbRates=sbRates.set_index('Region Name').T                                          #transposing the dataframe to desired place

sbRates                                                                             #outputting the dataframe
```

```python
plt.figure(figsize=(12,8))                                                          #defining the figure size

plt.title('Stillbirth Rate from 2010 to 2021 by UNICEF Region')                     #adding plot title

for region in sbRates:                                                              #plotting the data
    plt.plot(sbRates[region], label=region, marker='H')

plt.ylabel('Stillbirth Rate', fontsize=15)                                          #adding labels
plt.xlabel('Year', fontsize=15)

plt.legend(loc='upper center', bbox_to_anchor=(1.16, .8))                           #adding legends to the plot

plt.ylim(5000, 100000)                                                              #adjusting the tick marks
plt.show()
```

```python
exclusive_breastfeeding1 = pd.read_csv("datasets/exclusive_breastfeeding.csv", skiprows=2)

exclusive_breastfeeding1.rename(columns={'Unnamed: 1' : 'UNICEF Region', 'Unnamed: 2': 'Year', 'Unnamed: 3': 'Count'}, inplace=True)         #renaming columns
exclusive_breastfeeding1.at[5, 'UNICEF Region'] = 'NA'                                                                                       #renaming value
exclusive_breastfeeding1.at[8, 'Count'] = '0'  
exclusive_breastfeeding1.at[6, 'UNICEF Region'] = 'SAsia'                                                                                               #changing NaN value

exclusive_breastfeeding1 = exclusive_breastfeeding1.iloc[0:9]                                                                                 #slicing specific data
exclusive_breastfeeding1['Count'] = exclusive_breastfeeding1['Count'].astype(int)                                                             #converting column values to int

exclusive_breastfeeding1                                                                                                                     #displaying the dataframe
```

```python
plt.figure(figsize=(12, 8))

plt.bar(exclusive_breastfeeding1['UNICEF Region'], exclusive_breastfeeding1['Count'], color=['#fbf8cc', '#ffcfd2', '#cfbaf0', '#a3c4f3', '#90dbf4', '#8eecf5', '#98f5e1', '#b9fbc0'])

plt.title("Breastfeeding Aggregates via UNICEF Region")
plt.xlabel('UNICEF Region', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.grid(True)

plt.show()
```
